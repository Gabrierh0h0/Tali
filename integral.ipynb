```python
import numpy as np
from scipy import stats
from scipy.stats import poisson, expon, norm
import matplotlib.pyplot as plt
```

## Punto 3

**Contexto:** Se tienen datos de defectos y se asume una distribución de Poisson.

**Preguntas:**
1. Hallar el parámetro de la distribución.
2. ¿Cuál es la probabilidad de que tenga más de 5 defectos?
3. Hallar la variabilidad de los defectos.

**Método de Resolución:**
- Para la distribución de Poisson, el parámetro $\lambda$ equivale a la media de las observaciones o tasa de ocurrencia.
- La probabilidad de exceder un valor $k$ es $1 - P(X \leq k)$, lográndose mediante el método `poisson.cdf`.
- En Poisson, la varianza (variabilidad) es igual a $\lambda$.

```python
# Datos de entrada
datos_p3 = [8, 4, 4, 3, 4, 6, 4, 4, 5, 5, 4, 4, 5, 2, 3, 2, 2, 3, 8, 5]

# 1. Hallar el parámetro de la distribución (media)
lam = np.mean(datos_p3)
print(f"1. Parámetro de la distribución (λ): {lam:.2f}")

# 2. Probabilidad de que tenga más de 5 defectos P(X > 5)
k = 5
prob_mas_de_5 = 1 - poisson.cdf(k, lam)
print(f"2. Probabilidad de que tenga más de 5 defectos P(X > 5): {prob_mas_de_5:.4f} ({prob_mas_de_5*100:.2f}%)")

# 3. Variabilidad (Varianza en Poisson)
variabilidad = lam
print(f"3. Variabilidad de los defectos (Varianza): {variabilidad:.2f}")
```

## Punto 4

**Contexto:** Sea X la distancia con distribución exponencial y media de 72.15.

**Preguntas:**
1. Determinar el parámetro de la distribución.
2. ¿Cuál es la probabilidad de que la distancia sea como mucho 100m, como mucho 200m y entre 100 y 200m?
3. ¿Cuál es la probabilidad de que la distancia exceda la distancia media por más de 2 desviaciones estándar?

**Método de Resolución:**
- En la distribución Exponencial, la tasa $\lambda = 1 / \mu$. En Scipy se utiliza `scale = media` para resolverlo de manera directa usando `expon`.
- *"Como mucho x"* significa $P(X \leq x)$, calculado sumando el área bajo la curva con `expon.cdf(x)`.
- Por propiedades de la Exponencial, la desviación estándar ($\sigma$) es igual a la media ($\mu$).

```python
media_exp = 72.15

# 1. Determinar el parámetro de la distribución (λ = 1/Media)
lam_exp = 1 / media_exp
print(f"1. Parámetro de la distribución (λ): {lam_exp:.6f}")

# 2. Probabilidades: como mucho 100, como mucho 200, entre 100 y 200
prob_100 = expon.cdf(100, scale=media_exp)
prob_200 = expon.cdf(200, scale=media_exp)
prob_entre = prob_200 - prob_100

print(f"2a. Probabilidad (X <= 100m): {prob_100:.4f}")
print(f"2b. Probabilidad (X <= 200m): {prob_200:.4f}")
print(f"2c. Probabilidad (100m <= X <= 200m): {prob_entre:.4f}")

# 3. Exceder distancia media por más de 2 desviaciones estándar
# En dist. exponencial: σ = μ
desv_std = media_exp
limite = media_exp + 2 * desv_std
prob_excede = 1 - expon.cdf(limite, scale=media_exp)
print(f"3. Probabilidad de exceder la media en más de 2 σ (es decir, X > {limite:.2f}m): {prob_excede:.4f}")
```

## Punto 5

**Contexto:** Valores de concentración de cloruro y distribución Normal.

**Preguntas:**
1. Determinar si los datos siguen una distribución normal y justificar.
2. ¿Cuál es la probabilidad de que la concentración sea igual, menor y como mucho 105?
3. ¿Cuál es la probabilidad de que difiera de la media por más de una desviación estándar?
4. ¿Cómo caracterizaría el 0.1% más extremo de los valores de concentración de cloruro?

**Método de Resolución:**
- Para probar la normalidad usaremos el test de Shapiro-Wilk (si p-value > 0.05, los datos parecen normales).
- Se utiliza `norm.cdf()` para hallar las probabilidades de intervalos ($<$ o $\leq$), recordando que para continuas, la probabilidad puntual $P(X=x)$ es $0$.
- La diferencia de más de 1 $\sigma$ describe los extremos hacia arriba y abajo. Evaluado como $2 \cdot P(X \leq \mu - \sigma)$.
- Los extremos se examinan encontrando los límites interior (0.05%) y superior (99.95%) con la función `norm.ppf()` para abarcar el total del 0.1%.

```python
# Datos de concentración
datos_p5 = [104.3, 100.9, 99.0, 105.0, 113.8, 105.8, 
            102.2, 103.9, 98.5, 110.2, 94.5, 103.7, 
            92.6, 100.3, 107.4, 105.0, 101.6, 98.5, 
            105.3, 117.6, 113.0, 111.3, 105.6, 112.0, 
            104.0, 102.5, 104.5, 109.7, 96.1, 100.1]

mu_n = np.mean(datos_p5)
sigma_n = np.std(datos_p5, ddof=1)

print("1. Determinar normalidad:")
stat, p_valor = stats.shapiro(datos_p5)
print(f" - Prueba Shapiro-Wilk | Estadístico: {stat:.4f}, p-valor: {p_valor:.4f}")
if p_valor > 0.05:
    print(" - Como el p-valor > 0.05, no hay evidencia para rechazar la normalidad. (Justificado asumiendo distribución normal).\n")
else:
    print(" - Como el p-valor <= 0.05, se rechaza que provengan de una normal.\n")

print("2. Probabilidades respecto al valor 105:")
print(" - Igual a 105: 0.0000 (Las probabilidades de un punto exacto en distribución continua son 0)")
p_105 = norm.cdf(105, mu_n, sigma_n)
print(f" - Menor a 105 [P(X < 105)]: {p_105:.4f}")
print(f" - Como mucho 105 [P(X <= 105)]: {p_105:.4f} (Equivale al anterior por ser continua)\n")

print("3. Probabilidad de que difiera en más de 1 desviación estándar:")
dif_1_sigma = 2 * norm.cdf(mu_n - sigma_n, mu_n, sigma_n)
print(f" - P(|X - μ| > σ) = {dif_1_sigma:.4f} (aprox 31.73% clásico de la Normal)\n")

print("4. Caracterizar el 0.1% más extremo:")
limite_inferior = norm.ppf(0.0005, mu_n, sigma_n)
limite_superior = norm.ppf(0.9995, mu_n, sigma_n)
print(f" - El 0.1% de los valores más extremos corresponde a concentraciones menores a {limite_inferior:.2f} y concentraciones mayores a {limite_superior:.2f}.")
```
